```python
import numpy as np
import matplotlib.image as mpimg  # for reading in images
import matplotlib.pyplot as plt
import cv2 # computer vision library

%matplotlib inline
```

```python
# Read and Diplay the image
image = mpimg.imread('images/scene1.jpg')
print('Image Dimensions:', image.shape)
```

```python
# It's a color image with 3 layers: RGB
plt.imshow(image)
plt.show()
```

```python
# Change from color to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(16,10))
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(gray_image,cmap='gray')
plt.show()
```

```python
# The shapes of the images
print(image.shape)
print(gray_image.shape)

# from these pixels, select any specific value
x =150
y= 300

print(gray_image[y,x])
```

```python
# find the maximum and miniumu grayscale values in this image

max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print('Max pixel value:' ,max_val)
print('Min pixel value:', min_val)
```
